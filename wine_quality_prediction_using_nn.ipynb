```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.metrics import categorical_accuracy
```

```python
print(tf.__version__)
print(keras.__version__)
```

```python
def create_demomodel():
    model = Sequential()
#     first parameter of dense function is number of neurons or nodes in the layer.
    model.add(Dense(128, input_dim=11, activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(3, activation='softmax'))
    return model

wine_data = pd.read_csv('./Dataset/winequality-red.csv', sep=';')
# wine_data.loc[wine_data['quality'] > 6 : ]
wine_data.head()

data = wine_data.loc[wine_data['quality'] > 5, ].reset_index(drop=True)

data['quality'] = pd.Categorical(data['quality'])
y = pd.get_dummies(data['quality'], prefix='quality')

x = data.loc[:,list(data.columns)[:11]]
# y = data.loc[:,list(data.columns)[-1]]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)

```

```python
model = create_demomodel()
model.compile(optimizer='adam', loss='mse', metrics=['acc'])
model.fit(x=x_train, y=y_train, batch_size=10, epochs=6)
```

```python
prediction = model.predict(x=x_test, batch_size=10)
value = np.argmax(prediction, axis=1)
number_to_value = ['6_quality', '7_quality', '8_quality']
number = [0, 1, 2]
dic = dict(zip(number, number_to_value))

#evaluating model
scores = model.evaluate(x=x_test, y=y_test, verbose=1)
scores
for each in value:
    print(dic[each])
scores
```
